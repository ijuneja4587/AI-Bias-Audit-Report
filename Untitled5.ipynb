Approval Rate By Gender

```python
query_gender="""
SELECT gender, Loan_Approved, COUNT(*) * 1.0 / SUM(COUNT(*)) OVER(PARTITION BY Gender) As proportion
FROM df
GROUP BY Gender, Loan_Approved
"""
```

```python
gender_result = ps.sqldf(query_gender, locals())
print(gender_result)
```

```python
gender_pivot = gender_result.pivot(index='Gender', columns='Loan_Approved', values='proportion')
gender_pivot.plot(kind='bar', stacked=True, colormap='Pastel1')
plt.title('Loan Approval By Gender')
plt.ylabel('Proportion')
plt.show()
```

Criminal Record By Age Group & Gender

```python
query = """
SELECT Age_Group , Gender, COUNT(*) as Count
FROM df
WHERE Criminal_Record = 'Yes'
GROUP BY Age_Group, Gender
"""
```

```python
result = ps.sqldf(query, locals())
```

```python
print(result)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=result, x='Age_Group', y='Count', hue='Gender', palette='muted')
plt.title('Criminal Record Count by Age Group & Gender')
plt.xlabel('Age Group')
plt.ylabel('Criminal Records Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Average Credit Score By Employment Type

```python
query_avg_credit = """
SELECT Employment_Type, AVG(Credit_Score) as Avg_Credit_Score
FROM df
GROUP BY Employment_Type
"""
```

```python
avg_credit_result = ps.sqldf(query_avg_credit, locals())
```

```python
print(avg_credit_result)
```

```python
avg_credit_result.plot(kind='bar', x='Employment_Type', y='Avg_Credit_Score', color='orange', legend=False)
plt.title('Average Credit Score by Employment Type')
plt.ylabel('Average Credit Score')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
query_top_zip = """
SELECT Zip_Code_Group, AVG(Income) AS Avg_Income
FROM df
GROUP BY Zip_Code_Group
ORDER BY Avg_Income DESC
"""
```

```python
top_zip_result.plot(kind='bar', x='Zip_Code_Group', y='Avg_Income', color='Black', legend=False)
plt.title('Top Zip Code Groups by Average Income')
plt.ylabel('Average Income')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
